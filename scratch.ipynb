```python
logon = pd.read_csv("logon_info.csv")

print("Coverting Date")
logon['date'] = pd.to_datetime(logon['date'], format="%m/%d/%Y %H:%M:%S")
print("pickling")
logon.to_pickle("logon_info")
print("DONE")
```

```python
logon['id'] = logon['id'].astype('category')
logon['user'] = logon['user'].astype('category')
logon['pc'] = logon['pc'].astype('category')
logon['activity'] = logon['activity'].astype('category')

cat_columns = logon.select_dtypes(['category']).columns
logon[cat_columns] = logon[cat_columns].apply(lambda x: x.cat.codes)

logon
```

```python
g = sns.pairplot(logon, hue="activity")
```

```python
sns.distplot(logon['id'], rug=True, kde=True, hist=False, norm_hist=False)
```
