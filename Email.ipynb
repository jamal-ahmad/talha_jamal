# Sentiment Analysis

```python
email_sentiDict = dict()
users = emails["user"].unique()
for user in users:
    email_sentiDict[user] = (0, 0)
print "Dictioanry Length:\t", len(email_sentiDict), "\n"
pp.pprint(email_sentiDict)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize
import nltk
```

```python
nltk.download()
```

```python
from IPython.display import display  
from ipywidgets import FloatProgress 

f = FloatProgress(min=0,max=len(emails["content"]))  
display(f)
sid = SentimentIntensityAnalyzer()

for i in range(len(emails["content"])):
    f.value += 1
    old_cumulative, count = email_sentiDict[emails["user"][i]]
    count += 1
    ss = sid.polarity_scores(emails["content"][i])
    new_cummulative = old_cumulative + ss["compound"]
    email_sentiDict[emails["user"][i]] = (new_cummulative, count)
    
```

```python
import pprint as pp
pp.pprint(email_sentiDict)

import pickle as pkl
pickle_out = open("email_sentiDict.pickle","wb")
pkl.dump(email_sentiDict, pickle_out)
pickle_out.close()
```

```python
email_sentiDF = pd.DataFrame()
email_sentiDF = pd.DataFrame(email_sentiDict.items())
email_sentiDF.columns = ['user', 'score']
email_sentiDF[['senti', 'numEmails']] = email_sentiDF['score'].apply(pd.Series)
del email_sentiDF['score']
email_sentiDF.head()
```

```python
senti = email_sentiDF["senti"]
numEmails = email_sentiDF["numEmails"]
sentiPerEmail = np.divide(senti, numEmails)
email_sentiDF["sentiPerEmail"] = sentiPerEmail
email_sentiDF.head()
```

```python
email_sentiDF['user'] = email_sentiDF['user'].astype('category')

cat_columns = email_sentiDF.select_dtypes(['category']).columns
email_sentiDF[cat_columns] = email_sentiDF[cat_columns].apply(lambda x: x.cat.codes)
email_sentiDF.head()
```
